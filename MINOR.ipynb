```python
def catch_voice(speak_txt="Please tell the exercise name and also number of sets and reps for the exercise"):

    r=sr.Recognizer()
    engine.say(speak_txt)
    engine.runAndWait()
    with sr.Microphone() as source:
        audio=r.listen(source)
        
        try:
            text=r.recognize_google(audio)
            print("you said : {}".format(text))
        except:
            print("please speak again ,the audio couldn't be recorded")
            catch_voice()
            return

    ##if("yes" in text):
        ##find_variables(text)
    ##if("no" in text):
        ##calc_calories_burnt()
    
    return text
    
def speak(ins):
    engine.say(ins)
    engine.runAndWait()
```

```python
#code for finding no. of reps , sets and exercise name from the given text
def find_variables(text):
    reps=0
    sets=1
    exercise_name=""

    
    #text where number in words are replaced to number in digits
    for n,w in enumerate(units):

        if(text.find(w)!=-1):
            text=text.replace(w," {} ".format(n))

    print(text)
    
    #extracting no. of reps and sets using regular expression
    
    reg_ex=r"\d+\s\w+"
    re_list=re.findall(reg_ex,text)
    print(re_list)

    for x in re_list:
        reg_ex2=r"\d+"
        if(x.find('reps')!=-1):
            reps=re.findall(reg_ex2,x)[0]

        if(x.find('sets')!=-1):
            sets=re.findall(reg_ex2,x)[0]

    while(reps==0):
        rep_text=catch_voice("tell the number of reps you want to do")
        reg_ex2=r"\d+"
        rep_list=re.findall(reg_ex2,rep_text)
        print(rep_list)
        try:
            reps=rep_list[0]
        except:
            pass
    
    
    #for finding the name of the exercise
    for exer in exercises:
        if(text.find(exer)!=-1):
            exercise_name=exer
            break
    while(exercise_name==""):
        ex_text=catch_voice("tell the exercise name you want to do")
        reg_ex3=r"\w+"
        for exer in exercises:
            if(ex_text.find(exer)!=-1):
                exercise_name=exer
                break
        
        
    print("reps : {}".format(reps))
    print("sets : {}".format(sets))
    print("exercise_name : {}".format(exercise_name))
    
    engine.say("You are about to do {} sets of {} {}".format(sets,reps,exercise_name))
    engine.runAndWait()   #for output voice
    
    return (reps,sets,exercise_name)
    
```

```python

def calories(ExerciseName,reps):
    if ExerciseName == "biceps" or ExerciseName=='bicep': #curl biceps
        return (int(reps)*.1875) 
    elif ExerciseName == "lunges": 
        return (reps*0.3) 
    elif ExerciseName == 'squats': 
        return (reps*0.24) 
    elif ExerciseName == 'down': #latpulldown
        return (reps*0.8)
    
def time_func(st_time,end_time):
    total_sec=end_time-st_time
    return time.strftime("%H:%M:%S", time.gmtime(total_sec))
    

def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```
